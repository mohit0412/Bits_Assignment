#### Static Evaluation Value or Utility Code Design - Implement :

```python
import copy

'''
Function to check how many winning paths are there on current board position for a player. Its a helper function for
SEV calculation
'''
def count_winning_paths(player, board):
    count = 0
    # Check rows
    for i in range(5):
        for j in range(2):
            if all(board[i][j+k] == player for k in range(4)):
                count += 1
    # Check columns
    for i in range(2):
        for j in range(5):
            if all(board[i+k][j] == player for k in range(4)):
                count += 1
    # Check diagonals
    for i in range(2):
        for j in range(2):
            if all(board[i+k][j+k] == player for k in range(4)):
                count += 1
            if all(board[i+3-k][j+k] == player for k in range(4)):
                count += 1
    return count


'''
Helper function to fill all empty positions on board for a player. Its a helper function to calculate SEV
'''
def board_fill(brd, player):
    new_board = []
    for row in brd:
        new_row = []
        for square in row:
            new_row.append(square)
        new_board.append(new_row)

    for i in range(5):
        for j in range(5):
            if new_board[i][j] == '':
                new_board[i][j] = player
    return new_board


'''
Main function for implementation of static evaluation function
'''
def static_evaluation(board, max_player, min_player):
    """
    Evaluate the board state for the given players.

    Args:
        board (list): The current board state.
        max_player (str): The symbol used by the maximizing player.
        min_player (str): The symbol used by the minimizing player.

    Returns:
        int: The score of the board state for the maximizing player.
    """

    max_filled_board = board_fill(board, max_player)
    # Count the winning paths for the maximizing player
    max_score = count_winning_paths(max_player, max_filled_board)

    min_filled_board = board_fill(board, min_player)
    # Count the winning paths for the minimizing player
    min_score = count_winning_paths(min_player, min_filled_board)

    # Return the difference between the scores
#     print('SEV: ',max_score - min_score)
    return max_score - min_score
```

```python
'''
Helper function for min max algorithm.
'''
def get_possible_moves(board):
    """
    Returns a list of all possible moves in the current game state.
    Each move is represented as a tuple (row, col).
    """
    moves = []
    for i in range(5):
        for j in range(5):
            if board[i][j] == '':
                moves.append((i, j))
    return moves

'''
Helper function to make move
'''
def make_move(board, move, player):
    """
    Returns the new game state after making a move for the given player.
    """
    new_board = copy.deepcopy(board)
    i, j = move
    new_board[i][j] = player
    return new_board
```

### Terminal State / Game Ending (Win/Loss/Draw):  Implementation

```python
'''
Function to check whether the current state is a terminal state, i.e. any player's win or tie. Win is checked based on row, column,
two diagonals check.
'''
def terminal_state_check(board, player):
    # Check for a win or a tie
    win = False
    for row in board:
        row = [' ' if x == '' else x for x in row]
        if ''.join(row).strip().find(player*4) != -1:
            win = True
            break
    if not win:
        for i in range(5):
            column = [' ' if board[j][i] == '' else board[j][i] for j in range(5)]
            if ''.join(column).strip().find(player*4) != -1:
                win = True
                break
    if not win:
        d1 = [board[i][i] for i in range(len(board))]
        d2 = [board[i][len(board)-i-1] for i in range(len(board))]
        d1 = [' ' if x == '' else x for x in d1]
        d2 = [' ' if x == '' else x for x in d2]
        if ''.join(d1).strip().find(player*4) != -1 or ''.join(d2).strip().find(player*4) != -1:
            win = True

    if win:
        return 'win', player
    elif len(get_possible_moves(board)) == 0:
        return 'tie', player
    else:
        return 'running', player
```

### Implementation of the Min-Max algorithm

```python
'''
Main function for implementation of Min-Max algorithm, considering computer is max player.
'''
def minimax(board, player, depth):
    """
    Returns the best move and best score for the given player using the Minimax algorithm.
    """
    alpha_beta = {}
    if depth == 0:
        return None, static_evaluation(board, 'T', 'X'), alpha_beta
    if player == 'T':
        best_score = float('-inf')
        best_move = None
        for move in get_possible_moves(board):
            new_board = make_move(board, move, player)
            _, score, alpha_beta = minimax(new_board, 'X', depth-1)
            if score > best_score:
                best_score = score
                best_move = move
        alpha_beta['alpha'] = best_score
        return best_move, best_score, alpha_beta
    else:
        best_score = float('inf')
        best_move = None
        for move in get_possible_moves(board):
            new_board = make_move(board, move, player)
            _, score, alpha_beta = minimax(new_board, 'T', depth-1)
            if score < best_score:
                best_score = score
                best_move = move
        alpha_beta['beta'] = best_score
        return best_move, best_score, alpha_beta
```

```python
'''
Main function to play game.
'''
def play_game():
    # Initialize the empty board
    board = [['' for j in range(5)] for i in range(5)]

    # Choose the starting player (computer player goes first as max player)
    player = 'T'
    print('==================================')
    print('Game Started !!!')
    print('Initial Board State:')
    for i in board:
        print(i)
    print('Computer has coin : T\nHuman has coin : X')
    print('==================================')
    iteration = 1
    # Main game loop
    while True:
        print('Iteration : ', iteration)
        # Get the move for the current player
        if player == 'T':
            # Computer player's turn
            print("Computer's turn...")
            move, score, alpha_beta = minimax(board, player, depth=3)
            i, j = move
        else:
            # Human player's turn
            '''
            Dynamic input handling and throuwing error message if input is invalid.
            '''
            move = input("Your move (row, column) : ")
            while True:
                try:
                    i, j = map(int, move.split(','))
                    if 0 <= i < 5 and 0 <= j < 5 and board[i][j] == '':
                        break
                except:
                    pass
                move = input("Invalid move. Your move (row, column) (computer played first): ")

        # Make the move
        board[i][j] = player

        status, plr = terminal_state_check(board, player)
        if status == 'win':
            for b in board:
                print(b)
            print('Player {} won the game.'.format(plr))
            break
        elif status == 'tie':
            for b in board:
                print(b)
            print('Game stopped on tie.')
            break

        # Switch players
        player = 'X' if player == 'T' else 'T'

        # Print the board
        print("Current board:")
        for row in board:
            print(row)
        print('Alpha : ', alpha_beta['alpha'])
        print('Beta : ', alpha_beta['beta'])

        iteration += 1
        print('\n---------------------------------\n')
```
